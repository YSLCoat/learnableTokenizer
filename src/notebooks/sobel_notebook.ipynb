```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Create a blank 256x256 image with a medium gray background (intensity 150)
image = np.full((256, 256), 150, dtype=np.uint8)

# Draw a filled black square (high contrast) from (30, 30) to (150, 150)
cv2.rectangle(image, (30, 30), (150, 150), color=0, thickness=-1)

# Draw a larger filled rectangle with lower contrast (light gray, intensity 190)
cv2.rectangle(image, (90, 90), (230, 230), color=190, thickness=-1)

# Apply the Sobel operator to compute gradients in the x and y directions.
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)

# Compute the gradient magnitude (edge strength)
magnitude = np.sqrt(sobel_x**2 + sobel_y**2)
# Normalize the magnitude image for display purposes (scale to 0-255)
magnitude = np.uint8(255 * magnitude / np.max(magnitude))

# Plot the synthetic image and the gradient outputs.
plt.figure(figsize=(16, 4), dpi=200)

# Display the original synthetic image.
plt.subplot(1, 4, 1)
plt.imshow(image, cmap='gray')
plt.title('Input')
plt.axis('off')

# Display the Sobel X result.
plt.subplot(1, 4, 2)
plt.imshow(np.abs(sobel_x), cmap='gray')
plt.title('Gx')
plt.axis('off')

# Display the Sobel Y result.
plt.subplot(1, 4, 3)
plt.imshow(np.abs(sobel_y), cmap='gray')
plt.title('Gy')
plt.axis('off')

# Display the overall Edge Magnitude.
plt.subplot(1, 4, 4)
plt.imshow(magnitude, cmap='gray')
plt.title('Edge Magnitude')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Create a blank 256x256 image with a medium gray background (intensity 150)
image = np.full((256, 256), 150, dtype=np.uint8)

# Draw a filled black square (high contrast) from (30, 30) to (150, 150)
cv2.rectangle(image, (30, 30), (150, 150), color=0, thickness=-1)

# Draw a filled "white" square (actually light gray with intensity 190 for lower contrast)
# Same size (120x120) as the black square, positioned so the overlapping region is centered.
cv2.rectangle(image, (106, 106), (226, 226), color=190, thickness=-1)

# Apply the Sobel operator to compute gradients in the x and y directions.
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)

# Compute the gradient magnitude (edge strength)
magnitude = np.sqrt(sobel_x**2 + sobel_y**2)
magnitude = np.uint8(255 * magnitude / np.max(magnitude))  # Normalize for display

# Create separate figures with increased DPI for high resolution.

# Figure 1: Synthetic Image
plt.figure("Input", figsize=(6,6), dpi=200)
plt.imshow(image, cmap='gray')
plt.title('Input')
plt.axis('off')

# Figure 2: Sobel X
plt.figure("Sobel Gx", figsize=(6,6), dpi=200)
plt.imshow(np.abs(sobel_x), cmap='gray')
plt.title('Sobel Gx')
plt.axis('off')

# Figure 3: Sobel Y
plt.figure("Sobel Gy", figsize=(6,6), dpi=200)
plt.imshow(np.abs(sobel_y), cmap='gray')
plt.title('Sobel Gy')
plt.axis('off')

# Figure 4: Edge Magnitude
plt.figure("Edge Magnitude", figsize=(6,6), dpi=200)
plt.imshow(magnitude, cmap='gray')
plt.title('Edge Magnitude')
plt.axis('off')

plt.show()
```
